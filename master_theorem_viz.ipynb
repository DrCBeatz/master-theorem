```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the parameters for each algorithm
algorithms = {
    'Binary Search': {'a': 1, 'b': 2, 'f': lambda n: np.ones_like(n)},
    'Merge Sort': {'a': 2, 'b': 2, 'f': lambda n: n},
    'Strassen\'s Matrix Multiplication': {'a': 7, 'b': 2, 'f': lambda n: n**2}
}

# Define the range of n
n = np.linspace(1, 100, 400)

# Plotting
plt.figure(figsize=(18, 6))

for i, (name, params) in enumerate(algorithms.items(), 1):
    a, b, f = params['a'], params['b'], params['f']
    log_b_a = np.log2(a)
    n_log_b_a = n**log_b_a
    f_n = f(n)
    
    plt.subplot(1, 3, i)
    plt.plot(n, n_log_b_a, label='$n^{\\log_b a}$', color='blue')
    plt.plot(n, f_n, label='f(n)', linestyle='--', color='red')
    plt.xlabel('n')
    plt.ylabel('Value')
    plt.title(name)
    plt.legend()

plt.tight_layout()
plt.show()
```
